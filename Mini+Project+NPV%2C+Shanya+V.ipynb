```python
# 15. Draw a bar chart to show smoking and alcoholic habits against average expenditure
```

```python
b=pd.pivot_table(employee,values ='expenditure',index='alcohol',columns='smoke',aggfunc=np.mean)
b
```

```python
X=b.index
X_axis = np.arange(len(X))
Ysmoke=b['no']
Zsmoke=b['yes']
plt.bar(X_axis+0.2 , Ysmoke, 0.4, label = 'Non-Smokers')
plt.bar(X_axis -0.2, Zsmoke, 0.4, label = 'Smokers')
plt.xticks(X_axis,X)
plt.xlabel('Drinking Habits')
plt.ylabel('Average Expenditure')
plt.legend()
```

```python
# 16. Draw pie chart for alcohol consumption habits
```

```python
daily=employee[(employee['alcohol']=='daily')]
daily_c=daily['id'].count()
print ("Employees who drink daily", daily_c)

weekend=employee[(employee['alcohol']=='weekend')]
weekend_c=weekend['id'].count()
print ("Employees who drink on weekends", weekend_c)

party=employee[(employee['alcohol']=='party')]
party_c=party['id'].count()
print ("Employees who drink in party", party_c)

no=employee[(employee['alcohol']=='no')]
no_c=no['id'].count()
print ("Employees who dont drink", no_c)

rarely=employee[(employee['alcohol']=='rarely')]
rarely_c=rarely['id'].count()
print ("Employees who drink rarely", rarely_c)
```

```python
consumption=[daily_c, weekend_c, party_c, no_c, rarely_c]
l=['Daily', 'Weekend', 'Party', 'No', 'Rarely']
plt.pie(consumption, labels=l, autopct='%0.1f%%')
plt.show()
```
